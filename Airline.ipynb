# Cleaning

## Nan values

```python
df = df.dropna()
```

## Clean Flight Distance

```python
Q1 = df["Flight Distance"].quantile(0.25)
Q3 = df["Flight Distance"].quantile(0.75)
IQR = Q3 - Q1
df = df[
    ~(
        (df["Flight Distance"] < (Q1 - 1.5 * IQR))
        | (df["Flight Distance"] > (Q3 + 1.5 * IQR))
    )
]
```

## Departure Delay

```python
Q1 = df["Departure Delay"].quantile(0.25)
Q3 = df["Departure Delay"].quantile(0.75)
IQR = Q3 - Q1
df = df[
    ~(
        (df["Departure Delay"] < (Q1 - 1.5 * IQR))
        | (df["Departure Delay"] > (Q3 + 1.5 * IQR))
    )
]
```

## Arrival Delay

```python
df = df[df["Arrival Delay"] <= 20]
# custom value
```

# Visualisation

## Gender

```python
gender = df["Gender"].value_counts()
label = gender.index
color = ["#f5ff00", "#34499f"]
plt.pie(
    gender,
    labels=label,
    colors=color,
    wedgeprops={"edgecolor": "white", "linewidth": 3},
    autopct="%.2f",
)
plt.show()
```

## Age

```python
age = df["Age"]
bins = np.arange(10, 100, 10)
plt.hist(age, bins=bins, edgecolor="black")
plt.ylabel("Number of travelers")
plt.xlabel("Age")
plt.show()
```

- Most of the passengers are `40 - 50` years old.

## Class

```python
cls = df["Class"].value_counts()
plt.pie(
    cls,
    labels=cls.index,
    autopct="%.2f",
    wedgeprops={"edgecolor": "white", "linewidth": 2},
)
plt.show()
```

```python
classes = df["Class"].unique()
satisfied = []
disSatisfied = []
for c in classes:
    s = df[df["Class"] == c]["Satisfaction"].value_counts()
    satisfied.append(s["Satisfied"])
    disSatisfied.append(s["Neutral or Dissatisfied"])

fig, ax = plt.subplots()
ax.bar(classes, satisfied, label="Satisfied", width=0.5)
ax.bar(classes, disSatisfied, bottom=satisfied, label="Dissatisfied", width=0.5)
plt.title("Satisfaction of Travellers")
plt.xlabel("Class")
plt.ylabel("Number of Travellers")
ax.legend()
plt.show()
```

> most people in business class are satisfied but in economy and economy plus classes the majority are dissatisfied
